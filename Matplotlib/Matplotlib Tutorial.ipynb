```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=[1,2,3]
b=[2,4,6]

# scatter & line plot
plt.scatter(a,b)
plt.plot(a,b,'r-')
plt.show()
```

```python
x=np.arange(0,5,0.2)

# scatter
plt.plot(x,x**3,'ro',label='x**3')
plt.plot(x,x**2,'go',label='x**2')


plt.legend()
plt.axis([0,10,2,120])

plt.text(4,70,'text')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Matplotlib demo")
plt.show()
```

```python
# bubble graph

year=np.arange(2014,2020)
salary=np.arange(12,18)
population=np.random.randint(100,400,6)

plt.scatter(year,salary,s=population)
plt.title('Bubble Chart')
plt.show()

```

```python

# pie chart

a=np.random.randint(10,20,4)
print(a)
def f(p):
    if p>10 and p<20:return "All right"
    else:return "Not Right"
    
explode=[0.1,0,0,0] # should be less than 1
plt.pie(a,labels=['a','b','c','d'],explode=explode,autopct=f)
plt.show()
```

```python
# histogram

a=[1,1,2,2,3,3,4,4,4,4,5,10,12,10,19]
plt.xticks(np.arange(20))
plt.hist(a,bins=18,edgecolor='black')
plt.show()
```

```python
# bar graph

years=np.arange(2014,2019)
print(years)
population=np.random.randint(100,200,5)
print(population)
plt.bar(years,population)
plt.xticks(rotation=60)
plt.show()
```

**When to use which graph ?**

**Line graph is used to plot some Trend analysis.**
**Bar graph is used for trend analysis but in a small scale.**
**Scatter plot is used to find correlation betw two variables.**
**Pie chart is used when u want to show percentages and if the sum of the individual parts add up to a meaningful whole.**
**Bubble chart is used when 3 variables are there for analysis.**
**Histograms are used for frequency distribution*
