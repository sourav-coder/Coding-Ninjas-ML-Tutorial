```python
# 3

r=req.get("https://www.imdb.com/search/title/?release_date=2018&sort=num_votes,desc&page=1&ref_=adv_nxt")
data=BeautifulSoup(r.text)
li=data.find_all(class_='lister-item-content')
d={}
for i in li:
    votes=i.find_all(class_='sort-num_votes-visible')
    for j in votes:
        s=j.find_all('span')[1]['data-value']
        break
    d[i.a.string]=int(s)

d={i:j for i,j in sorted(d.items(),reverse=True,key=lambda x:x[1])}

s=''
m=-9999
for i in li:
#     print(i.a.string)
    ch=int(i.p.find(class_='runtime').string.split()[0])
#     print(ch)
    if ch>m:
        m=ch
        s=i.a.string
print(s,m)
print(f"{s} with {d[s]} votes")
    
```

```python
# 5


r=req.get("https://en.wikipedia.org/wiki/Artificial_intelligence")
data=BeautifulSoup(r.text)
m=-1
img=0
li=data.find_all('img')
for i in li:
    s=int(i['height'])*int(i['width'])
    if s>m:
        m=s
        img=i['src']
print(img,m)
```

```python
# 8
url="http://quotes.toscrape.com/"
r=req.get(url)
data=BeautifulSoup(r.text)

d={}
quotes=data.find_all(class_='quote')
for i in quotes:
    author=i.find_all('span')
    name=author[1].small.string
#     print(name)
    if name[0].lower()=='j':
        rq=req.get(url+author[1].a['href'])
        dt=BeautifulSoup(rq.text)
        dob=dt.find(class_='author-born-date').string
        d[name]=dob

d={i:j for i,j in sorted(d.items(),key=lambda x:x[0])}
print(d)
```
