### https://books.toscrape.com/


```python
url='https://books.toscrape.com/'
r=req.get(url)
data=BeautifulSoup(r.text)
```

```python
print(data.title.string)
```

```python
a=data.find(class_='product_pod')
a.h3.a['title'] # title of the first book
```

```python
# get all titles

titles=data.find_all(class_='product_pod')
for i in titles:
    print(i.h3.a['title'])
```

```python
# collect all urls
titles=data.find_all(class_='product_pod')
book_urls=[]
for i in titles:
    book_urls.append(url+i.h3.a['href'])
print(book_urls)

print(len(book_urls))
```

1. Get all images for all images of https://books.toscrape.com/

```python
sites=[]

for i in range(1,51):
    
    r=req.get(f'https://books.toscrape.com/catalogue/page-{i}.html')
    d=BeautifulSoup(r.text)
    titles=d.find_all(class_='product_pod')
    for i in titles:
        sites.append(url+i.h3.a['href'])
    
len(sites)
```
