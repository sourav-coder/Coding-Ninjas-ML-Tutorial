```python
# 2

setosa=df[df.flower_type=='Iris-setosa']
versicolor=df[df.flower_type=='Iris-versicolor']
virginica=df[df.flower_type=='Iris-virginica']

a=setosa.describe()
print(round(a.min().sw,2),round(a.min().sl,2),round(a.min().pl,2),round(a.min().pw,2),'Iris-setosa')
print(round(a.max().sw,2),round(a.max().sl,2),round(a.max().pl,2),round(a.max().pw,2),'Iris-setosa')
print(round(a.mean().sw,2),round(a.mean().sl,2),round(a.mean().pl,2),round(a.mean().pw,2),'Iris-setosa')

print()
a=versicolor.describe()
print(round(a.min().sw,2),round(a.min().sl,2),round(a.min().pl,2),round(a.min().pw,2),'Iris-versicolor')
print(round(a.max().sw,2),round(a.max().sl,2),round(a.max().pl,2),round(a.max().pw,2),'Iris-versicolor')
print(round(a.mean().sw,2),round(a.mean().sl,2),round(a.mean().pl,2),round(a.mean().pw,2),'Iris-versicolor')

print()
a=virginica.describe()
print(round(a.min().sw,2),round(a.min().sl,2),round(a.min().pl,2),round(a.min().pw,2),'Iris-virginica')
print(round(a.max().sw,2),round(a.max().sl,2),round(a.max().pl,2),round(a.max().pw,2),'Iris-virginica')
print(round(a.mean().sw,2),round(a.mean().sl,2),round(a.mean().pl,2),round(a.mean().pw,2),'Iris-virginica')
```
